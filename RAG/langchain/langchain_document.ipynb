```python
from langchain_community.document import Document

# create a Document object with some sample text
doc = Document(
                page_content="This is a sample document for testing purposes.",         
                metadata={"source": "sample.html",
                          "author": "santhi",
                          "date": "01-01-1994"
                        }
            )

print(doc.page_content)  # prints the text content of the document
```

```python
from langchain_community.document_loaders import TextLoader
doc = TextLoader("sample.txt").load()
print(f"Number of documents loaded: {len(doc)}")
print(f"Content of the first document: {doc[0].page_content}")  # prints the number of documents loaded and the text content of the first document
```
