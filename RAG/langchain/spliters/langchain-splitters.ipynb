```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
file_path = r"c:\\Users\\santhi.bhogavalli\\Downloads\\sample.txt"
data = TextLoader(file_path, encoding="utf-8").load()
chunck = CharacterTextSplitter(chunk_size=60, chunk_overlap=0, separator="\n")
#print(len(chunck.split_documents(data)))
split_documents = chunck.split_documents(data)
print(split_documents[0].page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
file_path = r"c:\\Users\\santhi.bhogavalli\\Downloads\\sample.txt"
loader = TextLoader(file_path, encoding="utf-8")
splitter = RecursiveCharacterTextSplitter(chunk_size=250, chunk_overlap=0)
split_documents = loader.load_and_split(splitter)
print(len(split_documents))
for i, doc in enumerate(split_documents):
    print(f"Document {i+1}:")
    print("-" * 40)
    print(doc.metadata)
    print("-" * 40)
    print(f"Content length: {len(doc.page_content)} characters")
    print(doc.page_content)
    print("-" * 40)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
import datetime
from langchain_text_splitters import RecursiveCharacterTextSplitter
file_path = r"c:\\Users\\santhi.bhogavalli\\Downloads\\sample.txt"
loader = TextLoader(file_path, autodetect_encoding=True)
chunk = RecursiveCharacterTextSplitter(chunk_size=50,separators=["\n", ""," "])
split = loader.load_and_split(chunk)
for metadata in split:
    metadata.metadata["project"] = "Learning LangChain"
    metadata.metadata["created_date"] = datetime.datetime.now().isoformat()
for i in range(len(split)):
    print(f"Document {i+1}:")
    print("-" * 40)
    print(split[i])
    print("-" * 40)
    print(split[i].metadata)
    print("-" * 40)
    print(f"Content length: {len(split[i].page_content)} characters")
    print(split[i].page_content)
    print("-" * 40)
```
